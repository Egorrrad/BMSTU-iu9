```julia
# Константы сопротивлений
const r12 = 1.0  # Ом
const r32 = 3.0  # Ом
const r24 = 1.0  # Ом
const r34 = 1.0  # Ом
const ξ   = 9.0  # В
```

```julia
# Формирование матрицы системы A
function create_matrix()
    A = zeros(3, 3)
    
    # Строка 1
    A[1, :] = [0, 0, 1]
    
    # Строка 2
    A[2, :] = [1, 0, -1]
    
    # Строка 3
    A[3, :] = [1/r12 + 1/r32, -(1/r12 + 1/r32 + 1/r24), 1/r24]
    
    return A
end

# Вектор правой части
b = [0, ξ, 0]

# Решение системы
A = create_matrix()
φ = A \ b  # φ (фи) - вектор потенциалов

# Вывод результатов
println("Матрица системы A:")
display(A)
println("\n")

println("Вектор правой части b:")
println(b)
println("\n")

println("Решение (потенциалы φ):")

println("φ[1] = $(φ[1])")
println("φ[2] = $(φ[2])")
println("φ[3] = $(φ[1])")
println("φ[4] = $(φ[3])")

# Вычисление токов
I12 = (φ[1] - φ[2]) / r12
I32 = (φ[1] - φ[2]) / r32
I24 = (φ[2] - φ[3]) / r24
I34 = (φ[1] - φ[3]) / r34


println("\nТоки:")
println("I12 = (φ1 - φ2)/r12 = $I12 A")
println("I32 = (φ3 - φ2)/r32 = $I32 A")
println("I24 = (φ2 - φ4)/r24 = $I24 A")
println("I34 = (φ3 - φ4)/r34 = $I34 A")
```
