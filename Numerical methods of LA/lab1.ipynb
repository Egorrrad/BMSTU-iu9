```julia
using PyPlot
# use x = linspace(0,2*pi,1000) in Julia 0.6
x = range(0; stop=2*pi, length=1000); y = sin.(3 * x + 4 * cos.(2 * x));
plot(x, y, color="red", linewidth=2.0, linestyle="--")
title("A sinusoidally modulated sinusoid")
```

```julia
using Gaston

N = 25
θ = range(0, 2π, length=N+1)
ϕ = range(0, 2π, length=N+1)

X = [(5 + 2 * cos(θ_i)) * cos(ϕ_j) for θ_i in θ, ϕ_j in ϕ]
Y = [(5 + 2 * cos(θ_i)) * sin(ϕ_j) for θ_i in θ, ϕ_j in ϕ]
Z = [2 * sin(θ_i) for θ_i in θ, ϕ_j in ϕ]

# Построение с сеткой
figure()
surf(X, Y, Z,
     title="Тор с сеткой",
     style="lines",
     linecolor="blue",
     linestyle="-",
     linewidth=0.7,
     box="on",
     grid="on")
```

```julia
using Gaston

t = range(0, 4π, length=6000)
X = (1 .+ 0.25 .* cos.(75 .* t)) .* cos.(t)
Y = (1 .+ 0.25 .* cos.(75 .* t)) .* sin.(t)
Z = t .+ 2.0 .* sin.(75 .* t)

# Создание цветового градиента на основе Z
colors = range(0, 1, length=length(Z))

figure()
plot3d(X, Y, Z,
      title="Параметрическая кривая с цветовой кодировкой",
      style="lines",
      linecolor=colors,
      palette="rainbow",
      linewidth=1.8,
      colorbar="on",
      colorbartitle="Высота Z")
```
